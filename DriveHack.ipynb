# Найдем точки притяжения

```python
#Создадим новый датасет и заполним нечисленные значения нулями
new_df = pd.read_csv("poezdki.csv")
new_df = new_df[["hash_ticket_uid", "start_trip_dttm", "start_latitude", "start_longitude", "start_entrance_nm", "start_station_nm", "start_line_nm", "start_stop_nm", "route_num", "vehicle_type", "end_trip_dttm", "end_station_nm", "end_line_nm", "end_stop_nm", "end_longitude", "trip_dur"]].fillna(0)
new_df
```

```python
#сделаем отдельный датасет только с автобусами
bus_df = new_df[['start_latitude', 'start_longitude', 'start_stop_nm', 'route_num', 'end_stop_nm', 'end_longitude']]
bus_df = bus_df.loc[bus_df['start_stop_nm'] != 0]
bus_df.head()
```

```python
#сделаем отдельный датасет только с метро
metro_df = new_df[['start_latitude', 'start_longitude','start_entrance_nm', 'start_station_nm', 'start_line_nm', 'end_station_nm', 'end_line_nm', 'end_longitude']]
metro_df = metro_df.loc[metro_df['start_entrance_nm'] != 0]
metro_df.head()
```

```python
#Отбираем из датасетов отдельные списки с координатами для карты и отбираем первые 10 для теста
bus_lat = bus_df['start_latitude']
bus_long = bus_df['start_longitude']
elevation_bus = bus_df.loc[df['start_stop_nm'] != 0]
elevation_bus = elevation_bus['start_stop_nm']

metro_lat = metro_df['start_latitude'] 
metro_long = metro_df['start_longitude']
elevation_metro = metro_df.loc[df['start_station_nm'] != 0]
elevation_metro = elevation_metro['start_station_nm']

bus_lat_test = bus_lat[:10]
bus_long_test = bus_long[:10]
elevation_bus_test = elevation_bus[:10]

metro_lat_test = metro_lat[:10]
metro_long_test = metro_long[:10]
elevation_metro_test = elevation_metro[:10]
```

```python
#зеленым отмечены станции метро, красным - автобусные остановки
bus_metro_map = folium.Map(location=[55.4507, 37.3656], zoom_start=8)

for bus_lat_test, bus_long_test, elevation_bus_test in zip(bus_lat_test, bus_long_test, elevation_bus_test):
    folium.Marker(location=[bus_lat_test, bus_long_test], popup=str(elevation_bus_test)+" station", icon = folium.Icon(color = 'red')).add_to(bus_metro_map)
                                                                                                                           
for metro_lat_test, metro_long_test, elevation_metro_test in zip(metro_lat_test, metro_long_test, elevation_metro_test):
    folium.Marker(location=[metro_lat_test, metro_long_test], popup=str(elevation_metro_test)+" station", icon = folium.Icon(color = 'green')).add_to(bus_metro_map)

bus_metro_map
```
